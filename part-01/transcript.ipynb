```
# Part 2: How to filter using boolean conditions
```

```
bool_cond = new_indexed_df.index.str.startswith('F')
another_cond = new_indexed_df.turnout > 70
new_indexed_df[bool_cond | another_cond]
```

```
cond_obama = new_indexed_df.winner == 'Obama'
new_indexed_df[cond_obama]
```

```
cond_romney = new_indexed_df.winner == 'Romney'
new_indexed_df[~cond_romney]
```

```
# Query Method
```

```
# turnout > 70
new_indexed_df.query('turnout > 70')
```

```
condition_one = 70
condition_two = 'F'
new_indexed_df.query('turnout > @condition_one & index.str.startswith(@condition_two)', engine='python')
```
